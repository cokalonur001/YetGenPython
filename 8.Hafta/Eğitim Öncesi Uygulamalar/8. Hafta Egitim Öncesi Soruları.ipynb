# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
def urunKaydet():
    while True:
        ad = input("Ürün adı giriniz (çıkmak için q tuşuna basın): ")
        if ad == "q":
            break
        fiyat = input("Ürün fiyatı giriniz: ")
        with open("urunler.txt", "a", encoding="utf-8") as file:
            file.write(ad + "," + fiyat + "\n")
        print(f"{ad} ürünü başarıyla kaydedildi.")
urunKaydet()
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
eski_kelime = input("Eski kelimeyi giriniz: ")
yeni_kelime = input("Yeni kelimeyi giriniz: ")

def kelimeDegistir(eski_kelime, yeni_kelime):

    with open("urunler.txt", "r", encoding="utf-8") as file:
        icerik = file.read()

    icerikDegisti = icerik.replace(eski_kelime, yeni_kelime)

    with open("urunler.txt", "w", encoding="utf-8") as file:
        file.write(icerikDegisti)
    print(f"{eski_kelime} kelimesi {yeni_kelime} kelimesiyle değiştirildi.")

kelimeDegistir(eski_kelime, yeni_kelime)
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, isim, hız, kilometre):
        self.isim = isim
        self.hız = hız
        self.kilometre = kilometre

    def bilgiGöster(self):
        print(f"Name: {self.isim} Speed: {self.hız} Mileage: {self.kilometre}")


class Bus(Vehicle):
    pass


school_bus = Bus("School Volvo", 180, 12)
school_bus.bilgiGöster()
````
