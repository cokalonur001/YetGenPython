# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
def urunKaydet():
    while True:
        ad = input("Ürün adı giriniz (çıkmak için q tuşuna basın): ")
        if ad == "q":
            break
        fiyat = input("Ürün fiyatı giriniz: ")
        with open("urunler.txt", "a", encoding="utf-8") as file:
            file.write(ad + "," + fiyat + "\n")
        print(f"{ad} ürünü başarıyla kaydedildi.")
urunKaydet()
````

1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız

````python
# 2.Problemin Çözümünü Buraya Yazınız
def kelimeDegistir(eski_kelime, yeni_kelime):
    eski_kelime =input("Eski Kelimeyi Giriniz: ")
    yeni_kelime =input("Yeni kelimeyi Giriniz: ")
    with open("urunler.txt", "r", encoding="utf-8") as file:
        icerik = file.read()

    icerik = icerik.replace(eski_kelime, yeni_kelime)

    with open("urunler.txt", "w", encoding="utf-8") as file:
        file.write(icerik)
````
