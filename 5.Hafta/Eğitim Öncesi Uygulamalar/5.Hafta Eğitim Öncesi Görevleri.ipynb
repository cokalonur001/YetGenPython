````python
# 1.Problemin Çözümünü Buraya Yazınız
genislik = float(input("Genişliği giriniz: "))
yukseklik = float(input("Yüksekliği giriniz: "))

def dikdortgenAlan(genislik, yukseklik):
    alan = genislik * yukseklik
    return alan
 
def dikdortgenCevre(genislik,yukseklik):
    cevre=(genislik+yukseklik)*2
    return cevre
 
print(f"Bahçenin Alanı: {dikdortgenAlan(genislik,yukseklik)}")
print(f"Bahçenin Çevresi: {dikdortgenCevre(genislik,yukseklik)}")
````
