# Problem 2 Faktöriyel

Kullanıcıdan bir sayı alınız ve alınan sayının faktöriyelini hesaplayan bir program yazınız. 


```python
# 2.Problemin Çözümünü Buraya Yazınız
sayi = int(input("Bir sayı girin: "))
faktoriyel = 1

if sayi < 0:
    print("Negatif sayıların faktöriyeli yoktur.")
elif sayi == 0:
    print("0'ın faktöriyeli 1'dir.")
else:
    for i in range(1, sayi + 1):
        faktoriyel = faktoriyel * i
    print(f"{sayi}! = {faktoriyel}")
```

# Problem 3 Virgül ile Ayırma

Kullanıcıdan bir metin halinde virgül ile ayrılmış sayılar alınız ve aldığınız her bir sayıyı listeye atayınız, daha sonra listedeki her elemanı toplayıp ekrana bastıran bir program yazınız.

<h4>Sample Input</h4>
<p>
Lutfen sayilari virgul ile ayrirarak giriniz: 5,14,7,2,1
</p>
<h4>Output</h4>

Toplam: 29

```python
# 3.Problemin Çözümünü Buraya Yazınız
sayilar_metin = input("Virgülle ayrılmış sayıları girin: ")
sayilar = sayilar_metin.split(",")
toplam = 0

for sayi in sayilar:
    toplam += int(sayi)

print(f"Girilen sayıların toplamı: {toplam}")
```

# Problem 4 Büyük Sayı

Kullanıcıdan alınan iki sayıdan büyük olanı bulup, ekrana yazdıran bir fonksiyon yazınız.

```python
# 4.Problemin Çözümünü Buraya Yazınız
def buyuk_bul(sayi1, sayi2):
    if sayi1 > sayi2:
        print(f"{sayi1} büyüktür.")
    elif sayi1 < sayi2:
        print(f"{sayi2} büyüktür.")
    else:
        print("Girilen sayılar birbirine eşittir.")

# Fonksiyonu test etmek için:
buyuk_bul(5, 10)
buyuk_bul(100, 50)
buyuk_bul(7, 7)
```

# Problem 5 Ürün Kayıt Uygulaması
Programımız ilk olarak kullanıcıdan, kaydı yapılacak ürünlerin sayısını isteyecek. Ardından yazılmış olan ürün sayısı kadar ürün ismi ve ürün fiyatı bilgisini kullanıcıdan alacak. Son olarak ise ürünleri ekrana yazdıracağız.

<h4>Sample Input:</h4>
<p>
Lutfen kac adet urun gireceginizi yaziniz: 3
<br/>
1. Urununuzun ismini giriniz: Mentos
<br/>
1. Urununuzun fiyatini giriniz: 1
<br/>
2. Urununuzun ismini giriniz: Meybuz
<br/>
2. Urununuzun fiyatini giriniz: 0.25
<br/>
3. Urununuzun ismini giriniz: Uzun Şeker
<br/>
3. Urununuzun fiyatini giriniz: 0.75
<br/>
<h4>Output:</h4>
1. Urun: Mentos - 1 TL
<br/>
2. Urun: Meybuz - 0.25 TL
<br/>
3. Urun: Uzun Şeker - 0.75 TL
</p>

```python
# 5.Problemin Çözümünü Buraya Yazınız
urun_sayisi = int(input("Kaç adet ürün kaydedeceksiniz? "))
urunler = {}

for i in range(urun_sayisi):
    urun_adi = input(f"{i+1}. ürünün adı: ")
    urun_fiyati = float(input(f"{urun_adi} fiyatı: "))
    urunler[urun_adi] = urun_fiyati

print("Kaydedilen ürünler:")
for urun, fiyat in urunler.items():
    print(f"{urun}: {fiyat} TL")
```

# Problem 6 Türkçe Karakter Çevirici
Programa girilen dosya ismindeki Türkçe karakterleri, İngilizce'ye çeviren programı yazınız.

<h4>Sample Input:</h4>
<p>
Lutfen dosya ismi giriniz: fotoğraf.jpeg
<br>
Dosya ismi: fotograf.jpeg
<br>
<br>
Lutfen dosya ismi giriniz: ışık.png
<br>
Dosya ismi: isik.png
</p>

```python
# 6.Problemin Çözümünü Buraya Yazınız
import os
char_dict = {'ı': 'i', 'ö': 'o', 'ü': 'u', 'ç': 'c', 'ğ': 'g', 'ş': 's', 'İ': 'I', 'Ö': 'O', 'Ü': 'U', 'Ç': 'C', 'Ğ': 'G', 'Ş': 'S'}
dosya_ismi = input("Dosya ismini girin: ")
if os.path.isfile(dosya_ismi):
    with open(dosya_ismi, "r", encoding="utf-8") as dosya:
        icerik = dosya.read()
        for turkce, ingilizce in char_dict.items():
            icerik = icerik.replace(turkce, ingilizce)
    with open("cevrilmis_" + dosya_ismi, "w", encoding="utf-8") as yeni_dosya:
        yeni_dosya.write(icerik)
    print(f"{dosya_ismi} dosyasındaki Türkçe karakterler İngilizce karakterlere çevrilerek 'cevrilmis_{dosya_ismi}' adlı yeni bir dosya oluşturuldu.")
else:
    print(f"{dosya_ismi} adlı dosya bulunamadı.")
```
